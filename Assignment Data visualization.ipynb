## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
stores=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
Clothes=np.array([82,78,48,46,39])
Equipments=np.array([48,31,51,29,18])
Accessories=np.array([15,20,14,13,9])
out=[x for x, _ in enumerate(stores)]
plt.bar(out,Clothes,width=0.8,label='Clothing',color='yellow',bottom=Equipment+Accessories)
plt.bar(out,Equipments,width=0.8,label='Equipments',color='red',bottom=Accessories)
plt.bar(out,Accessories,width=0.8,label='Accessories',color='purple')
plt.xticks(out,stores,rotation=90)
plt.xlabel("Cities")
plt.ylabel("Stores")
plt.legend(loc='upper right')
plt.title("Stores in various cities")
```
